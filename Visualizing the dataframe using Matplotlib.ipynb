```python
#To arrange the data into the axis
fig, ax = plt.subplots(2, 2)
data['Open'].plot(ax=ax[0, 0], title="Open")
data['High'].plot(ax=ax[0, 1], title="High")
data['Low'].plot(ax=ax[1, 0], title="Low")
data['Close'].plot(ax=ax[1, 1], title="Close")
```

```python
#Its still inadequate because the dates are not arranged
#To arrange the dates
fig, ax = plt.subplots(2, 2)
data['Open'].plot(ax=ax[0, 0], title="Open")
data['High'].plot(ax=ax[0, 1], title="High")
data['Low'].plot(ax=ax[1, 0], title="Low")
data['Close'].plot(ax=ax[1, 1], title="Close")
plt.tight_layout()
```

```python
# To show the Volume of stocks within a given interval (say, 2021-05-07 to 2021-06-06) using a bar chart
fig, ax = plt.subplots()
data['Volume'].loc['2021-05-07':'2021-06-06'].plot.barh(ax=ax)
```

```python
#To make the chart vertical
fig, ax = plt.subplots()
data['Volume'].loc['2021-05-07':'2021-06-06'].plot.bar(ax=ax)
```
